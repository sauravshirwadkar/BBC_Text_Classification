```python
import pandas as pd
# Dataset source : https://www.kaggle.com/code/rahulkumarpatro/multi-class-text-classification-using-lstms/input
df = pd.read_csv("bbc-text.csv")
df.head()
```

```python
df.shape
```

```python
df.category.value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.heatmap(df.isnull())
```

```python
df["target_encoded"] = df.category.replace({
    "sport" : 0,    
    "business" : 1,       
    "politics" : 2,       
    "tech" : 3,           
    "entertainment" : 4 
})
df.head(10)
```

```python
import spacy
nlp = spacy.load("en_core_web_lg")
```

```python
df["vector"] = df["text"].apply(lambda x : nlp(x).vector)
```

```python
df.head()
```

```python
df["vector"][0].shape
```

### import torch
import numpy as np
def ndarray_to_tensor(ndarray):
    if isinstance(ndarray, np.ndarray):
        return torch.from_numpy(ndarray)
    else:
        raise ValueError("Input must be a NumPy ndarray")

```python
df["tensor_vector"] = df["vector"].apply(ndarray_to_tensor)
```

```python
df.head()
```

```python
df["tensor_vector"][0]
```

```python
import tensorflow as tf
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from tensorflow.keras.layers import Conv1D
from tensorflow.keras.layers import MaxPooling1D
from tensorflow.keras.layers import Embedding, Flatten
from keras.preprocessing import sequence
```

```python
from sklearn.model_selection import train_test_split
X = df.tensor_vector.values
y = df.target_encoded
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, random_state = 42)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
# Padding the data samples to a maximum review length in words
max_words = 300
X_train = sequence.pad_sequences(X_train, maxlen=max_words)
X_test = sequence.pad_sequences(X_test, maxlen=max_words)
# Building the CNN Model
model = Sequential()      # initilaizing the Sequential nature for CNN model
# Adding the embedding layer which will take in maximum of 450 words as input and provide a 32 dimensional output of those words which belong in the top_words dictionary
model.add(Flatten(input_shape=(300,)))
model.add(Embedding(300, 32, input_length=max_words))
model.add(Conv1D(32, 3, padding='same', activation='relu'))
model.add(MaxPooling1D())
model.add(Flatten())
model.add(Dense(250, activation='relu'))
model.add(Dense(1, activation='softmax'))
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# Fitting the data onto model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=2, batch_size=128, verbose=2)
```
